# Exploratory data analysis

```python
# Getting month year from order_date
data['month_year'] = data['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
# grouping month_year by sales
data_temp = data.groupby('month_year').sum()['sales'].reset_index()
```

```python
plt.figure(figsize=(16, 5))
plt.plot(data_temp['month_year'], data_temp['sales'], c="b")
plt.xticks(rotation='vertical', size=8)
plt.show()
```

# THE TOP 10 PRODUCTS BY SALES

```python
# Grouping products by sales
sales = pd.DataFrame(data.groupby('product_name').sum()['sales'])

# Sorting the dataframe in descending order
sales.sort_values(by=['sales'], inplace=True, ascending=False)

# Top 10 products by sales
sales[:10]
```

# THE MOST SELLING PRODUCTS

```python
# Grouping products by Quantity
selling_prods = pd.DataFrame(data.groupby('product_name').sum()['quantity'])

# Sorting the dataframe in descending order
selling_prods.sort_values(by=['quantity'], inplace=True, ascending=False)

# Most selling products
selling_prods[:10]
```
